```python
import numpy as np
```

```python
#buat vector x dengan ukuran 3x1
x = np.array([12, 12, 12])
```

```python
#buat vector x dengan ukuran 3x1
x
```

```python
#hitung nilai L2 norm dari vector x
np.linalg.norm(x)
```

```python
#buat vector v dengan ukuran 3x1
v = np.array([3, 4])
```

```python
#buat vector v dengan ukuran 3x1
v
```

```python
#hitung nilai L2 norm dari vector v
np.linalg.norm(v)
```

```python
#tampilkan unit vector dari vector v
norm = np.linalg.norm(v)
```

```python
#tampilkan unit vector dari vector v
unit_v = v / norm
print("Unit vector:", unit_v)
```

```python
#hitung panjang unit vector dari vector v
np.linalg.norm(unit_v)
```
